**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
dfgm.head()
```

```python
def create_price_ratio(borough): 
    # filter for 1998 and 2018
    y1998 = dfgm.loc[(dfgm['london_borough'] == borough) & (dfgm['year'] == 1998), 'average_price'].mean()
    y2018 = dfgm.loc[(dfgm['london_borough'] == borough) & (dfgm['year'] == 2018), 'average_price'].mean()
    
    ratio = y1998 / y2018
    return ratio
```

```python
# get boroughs:
boroughs = dfgm['london_borough'].unique()

# iterate
for b in boroughs:
    ratio = create_price_ratio(b)
    print(f"Ratio for {b}: {ratio}")
```

### Initial Breif:
*Which boroughs of London have seen the greatest house price increase, on average, over the past two decades?*

The answer to the initial question is that **Hackney** is the borough that has had the largest increase in house prices in the past two decades. When the house prices from 1998 are divided by 2018, the ratio is smaller than any other borough, meaning house prices grew the most. A smaller ratio means more change.

```python
# find borough with the smallest ratio:
ratio2 = [create_price_ratio(b) for b in boroughs]

smallest_ratio = min(ratio2)
sr_index = ratio2.index(smallest_ratio)
borough_w_sr = boroughs[sr_index]

print(f"The smallest ratio is {smallest_ratio} for the borough {borough_w_sr}")
```

I could graph the prices of each borough but it would be difficult to campare all 32 of them. So I graphed the boroughs with the most change, the least change, and the median amount of change.

```python
# find borough w largest ratio and median ratio:
import statistics as s

largest_ratio = max(ratio2)
median_ratio = s.median(ratio2)

lr_index = ratio2.index(largest_ratio)
borough_w_lr = boroughs[lr_index]

mr_index = ratio2.index(median_ratio)
borough_w_mr = boroughs[mr_index]

print(f"The largest ratio is {largest_ratio} for the borough {borough_w_lr}")
print(f"The median ratio is {median_ratio} for the borough {borough_w_mr}")
```

```python
# make var of prices of a particular borough
hounslow_prices = df[df['london_borough'] == 'Hounslow']
merton_prices = df[df['london_borough'] == 'Merton']
hackney_prices = df[df['london_borough'] == 'Hackney']

# make line plot
plt.plot(hackney_prices['month'], hackney_prices['average_price'], label='Hackney')
plt.plot(merton_prices['month'], merton_prices['average_price'], label='Merton')
plt.plot(hounslow_prices['month'], hounslow_prices['average_price'], label='Hounslow')

# labels 
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.title('Average Housing Prices in Selected Boroughs')
plt.legend()

plt.show()
```
